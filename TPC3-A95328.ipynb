```python

def menu ():
    print ("""Gestão de listas de inteiros
              
              (1) Introduza a lista à mão;
              (2) Gerar uma lista com valores aleatórios;
              (3) Máximo da lista;
              (4) Organizar a lista;
              (0) Sair.""")

def criarListaM():
    n=int(input("Introduza o nº de elementos da lista: "))
    L=[]
    for i in range (n):
        elem=int(input("Introduz o elemento"+ str(i+1)+":"))
        L.append(elem)
    return L

def criarListaA():
    import random
    n=int(input("Introduza o nº de elementos da lista: "))
    L=[]
    for i in range (n):
        elem=random.randint (0,100)
        L.append(elem)
    return L

def maiorElemento(lista):
    maior=lista[0]
    i=1
    while (i<= len(lista)-1):
        if lista[i]>maior:
            maior=lista[i]
        i=i+1
    return maior

def BubbleSort(lista):
    trocas=1 #contador de trocas que o cíclo se vai basear para dar o término da função
    while (trocas != 0):
        i=0
        while (i <= len(lista)-2):
            if(lista[i]>lista[i+1]):
                troca1=lista[i]
                troca2=lista[i+1]
                lista[i+1]=troca1
                lista[i]=troca2
                trocas+=1
            i+=1
        trocas-=1 #como trocas começa em 1 temos de retirar 1 no fim de cada cíclo para ter o valor real das trocas  que ocorreram
    return lista

op="1"
lista=[]
while (op!="0"):
    menu()
    op=input("Qual a sua opção:")
    if ((op=="3" or op=="4") and len(lista)==0):
        print("Antes de usar esta opção crie uma lista.")
    elif (op=="1"):
        lista=criarListaM()
        print (lista)
    elif (op=="2"):
        lista=criarListaA()
        print (lista)
    elif (op=="3" and len(lista)!=0):
        maior=maiorElemento(lista)
        print (maior)
    elif (op=="4" and len(lista)!=0):
        lista_org = BubbleSort(lista)
        print(lista_org)
    elif (op=="0"):
        print ("Adeus!")
    else: 
        print ("Escolha outra opção!")
        
        
```
